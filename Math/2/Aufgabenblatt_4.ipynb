### Aufgabe 1

Bestimmen Sie folgenden Grenzwert:

$$
\lim_{x \rightarrow 4} \frac{x-4}{x^2-x-12}
$$

Plotten Sie die Funktion.

$$
\lim_{x \rightarrow 4} \frac{x-4}{(x+3)(x-4)}
$$

$$
\lim_{x \rightarrow 4} \frac{1}{(x+3)}
$$

Ersetzen wir x

$$
\lim_{x \rightarrow 4}\frac{1}{4+3}
$$

$$
\lim_{x \rightarrow 4}\frac{1}{7}
$$

$$
\frac{1}{7}
$$

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x):
   return (x-4)/(x**2-x-12)
def g(z):
   return (z-4)/(z**2-z-12)
x = np.linspace(-10, -3, 1000)
z = np.linspace(-3, 10, 1000)
plt.ylim(-20, 20)
plt.axhline()
plt.axvline()
plt.axhline(color='black', lw=0.5)
plt.axvline(color='black', lw=0.5)
plt.plot(x, f(x), color='blue')
plt.plot(z, g(z), color='blue')
plt.grid()
plt.show()
```
