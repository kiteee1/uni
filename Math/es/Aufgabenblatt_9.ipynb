### Aufgabe 1

Um drei Trainingsmethoden A, B und C für den Speerwurf zu erproben, wurden 23 untrainierte Sportstudenten zufällig in drei Gruppen zu $n_1=6$, $n_2=9$ und $n_3=8$ Studenten eingeteilt. Vor Beginn der Trainingsphase wurde zunächst ein Leistungstest durchgeführt und für jeden Studenten die Weite des besten von drei Würfen notiert.

Nach Abschluss der Trainingsphase, während der die Studenten der Gruppe 1 nach Methode A, die der Gruppe 2 nach Methode B und die der Gruppe 3 nach Methode C trainiert wurden, wurde ein entsprechender Leistungstest durchgeführt. Es ergaben sich die folgenden Werte (Differenzen der ermittelten Werte beim ersten und zweiten Leistungstest in Meter):

$$
\begin{array}{|l|ccccccccc|}
  \hline
  % after \\: \hline or \cline{col1-col2} \cline{col3-col4} ...
  \textrm{Gruppe 1} & 7.06 & 13.50 & 4.86 & 12.00 & 8.38 & 9.20 &  &  &  \\ \hline
  \textrm{Gruppe 2} & 8.73 & 10.24 & 9.78 & 7.30 & 7.42 & 10.54 & 12.98 & 17.80 & 9.40 \\ \hline
  \textrm{Gruppe 3} & 5.00 & 10.88 & 4.70 & 6.90 & 12.80 & 8.68 & 8.34 & 9.04 &  \\
  \hline
\end{array}
$$

Es wird angenommen, dass die ermittelten Werte eine Realisierung von unabhängigen in der Gruppe $i$ identisch $N(\mu_i,\sigma^2)$-verteilten Zufallsvariablen sind $(i=1,2,3)$. Unter diesen Annahmen teste man mit Hilfe der einfachen Varianzanalyse zum Niveau 5\% die Annahme: "Der durchschnittliche Trainingserfolg ist bei allen Trainingsmethoden gleich".

Die Nullhypothese lautet:

$$
H_0:\quad \text{„Der mittlere Trainingserfolg ist in allen drei Gruppen gleich.“}
$$

Die Alternativhypothese besagt, dass sich mindestens ein Gruppenmittelwert unterscheidet:

$$
H_1:\quad \text{„Mindestens ein Gruppenmittelwert unterscheidet sich.“}
$$

## 1. Gruppensummen und Gruppenmittelwerte

$$

\overline{x}_3

$$

$$
\overline{x} =\frac{T}{n}

### Aufgabe 2

In einer Fabrik wurden innerhalb eines Jahres 100 Fälle registriert, in denen ein Arbeitnehmer genau einen Tag bei der Arbeit fehlte. Davon entfielen auf die einzelnen Wochentage

$$
\begin{array}{|l|c|c|c|c|c|}
  \hline
  % after \\: \hline or \cline{col1-col2} \cline{col3-col4} ...
  \textrm{Wochentag} & \textrm{Mo} & \textrm{Di} & \textrm{Mi} & \textrm{Do} & \textrm{Fr} \\ \hline
  \textrm{Anzahl} & 22 & 19 & 16 & 18 & 25 \\
  \hline
\end{array}
$$

Ist die Annahme haltbar, dass sich solche eintägigen Arbeitsausfälle gleichmäßig auf die fünf Arbeitstage verteilen? Man prüfe eine entsprechende Hypothese mit dem $\chi^2$-Anpassungstest zum Niveau 5\%.

**Chi-Quadrat-Anpassungstest zum Niveau 5%**

Wir haben in einer Fabrik innerhalb eines Jahres 100 eintägige Arbeitsausfälle beobachtet, die sich auf die fünf Arbeitstage wie folgt verteilen:

- Montag:  22
- Dienstag: 19
- Mittwoch: 16
- Donnerstag: 18
- Freitag: 25

Wir möchten prüfen, ob diese Beobachtungen **gleichmäßig** über die fünf Wochentage verteilt sind. Die **Nullhypothese** lautet:

$$
H_0: \text{Die Verteilung der Ausfälle ist gleichmäßig, d.h. } p_1 = p_2 = p_3 = p_4 = p_5 = \frac{1}{5}.
$$

Die **Alternativhypothese** besagt, dass die Verteilung nicht gleichmäßig ist.

## 1. Erwartete Häufigkeiten

Unter ($H_0$) wäre der jeweilige Anteil für jeden Tag $\tfrac{1}{5}$. Die **erwartete** Häufigkeit in jeder Kategorie (Wochentag) beträgt somit:

$$
\text{Erwartete Häufigkeit} 
= \frac{1}{5} \times 100 
= 20.
$$

Wir fassen die Daten in einer Tabelle zusammen:

| Wochentag | Beobachtet \(O_i\) | Erwartet \(E_i\) |
|-----------|--------------------|-------------------|
| Mo        | 22                 | 20               |
| Di        | 19                 | 20               |
| Mi        | 16                 | 20               |
| Do        | 18                 | 20               |
| Fr        | 25                 | 20               |
| **Summe** | 100                | 100              |

## 2. Teststatistik

Der **Chi-Quadrat-Anpassungstest** verwendet folgende Teststatistik:

$$
\chi^2 
= \sum_{i=1}^{5} \frac{(O_i - E_i)^2}{E_i}.
$$

Wir berechnen termweise:

- Montag: (22 - 20)^2 / 20 = 4 / 20 = 0,20
- Dienstag: (19 - 20)^2 / 20 = 1 / 20 = 0,05
- Mittwoch: (16 - 20)^2 / 20 = 16 / 20 = 0,80
- Donnerstag: (18 - 20)^2 / 20 = 4 / 20 = 0,20
- Freitag: (25 - 20)^2 / 20 = 25 / 20 = 1,25

Zusammengezählt:

$$
\chi^2 
= 0,20 + 0,05 + 0,80 + 0,20 + 1,25 
= 2,50.
$$


## 3. Freiheitsgrade und kritischer Bereich

Wir haben 5 Kategorien (Wochentage) und die Hypothese, dass jede Kategorie denselben Anteil bekommt. Für den Chi-Quadrat-Anpassungstest lautet die **Freiheitsgradzahl**:

$$
\text{df} = k - 1 = 5 - 1 = 4.
$$

Auf dem **5%-Niveau** (Signifikanzniveau \(\alpha = 0{,}05\)) ist der **kritische Wert**:

$$
\chi^2_{0,05; (4)} 
\approx 9,488.
$$

Weil unsere berechnete Teststatistik $\chi^2 = 2,50$ kleiner ist als $9,488$, **verwerfen** wir $H_0$ **nicht**.

## 4. Fazit

Da

$$
\chi^2_\text{beob} 
= 2,50 
< 
\chi^2_{0,05;(4)} 
\approx 9,488
$$

liegt der **p-Wert** über 5 %. Wir haben **keinen** statistisch signifikanten Hinweis auf eine Abweichung von einer gleichmäßigen Verteilung. Die Daten sind **vereinbar** mit der Annahme, dass die Ausfälle gleichmäßig über die fünf Arbeitstage verteilt sind.


### Aufgabe 3

Unter einem Mikroskop wurden nacheinander 650 disjunkte Ausschnitte aus einer Zellkultur beobachtet und jeweils die Anzahl der Zellen bestimmt, die sich in einer bestimmten Phase der Zellteilung befinden. Die folgende Tabelle gibt an, wie oft ($y_i$-mal) bei den 650 Beobachtungen genau $i$ solcher Zellen registriert wurden.

$$
\begin{array}{|c|c|c|c|c|c|c|c|}
  \hline
  % after \\: \hline or \cline{col1-col2} \cline{col3-col4} ...
  i & 0 & 1 & 2 & 3 & 4 & 5 & 6 \\ \hline
  y_i & 72 & 169 & 191 & 120 & 66 & 21 & 11 \\
  \hline
\end{array}
$$

Man teste mit Hilfe des $\chi^2$-Anpassungstests zum Niveau $\alpha=5\%$ die Hypothese, dass die Anzahl solcher Zellen in einem Ausschnitt durch eine Poisson-verteilte Zufallsvariable beschrieben werden kann.


**Chi-Quadrat-Anpassungstest (Goodness-of-Fit) zum Niveau $\alpha = 5%$

In einem Experiment wurden $650$ disjunkte „Ausschnitte“ (Beobachtungen) aus einer Zellkultur untersucht. Dabei zählte man jeweils die Anzahl von Zellen in einer bestimmten Phase der Zellteilung. Die beobachteten Häufigkeiten $y_i$ für $i=0,1,\dots,6$ Zellen in einem Ausschnitt sind:

$$
\begin{array}{c|ccccccc}
i      & 0   & 1   & 2   & 3   & 4   & 5   & 6 \\
\hline
y_i    & 72  & 169 & 191 & 120 & 66  & 21  & 11
\end{array}
$$
$$
\text{(Summe über alle } i:  72+169+191+120+66+21+11 = 650).
$$

Wir wollen testen, ob diese Zufallsvariable (die Anzahl der Zellen im Ausschnitt) **Poisson-verteilt** ist.

## 1. Schätzung des Poisson-Parameters \(\lambda\)

Bei einer Poisson-Verteilung mit Parameter $\lambda$ ist der **Erwartungswert** $\lambda$ gleich dem **Stichprobenmittel**. Wir berechnen daher zunächst

$$
\widehat{\lambda}
= \frac{\sum_{i=0}^6 i \, y_i}{\sum_{i=0}^6 y_i}
= \frac{0\cdot72 + 1\cdot169 + 2\cdot191 + 3\cdot120 + 4\cdot66 + 5\cdot21 + 6\cdot11}{650}.
$$

$$1\cdot169 = 169$$

$$2\cdot191 = 382$$

$$3\cdot120 = 360$$

$$4\cdot66  = 264$$

$$5\cdot21  = 105$$

$$6\cdot11  = 66$$

Somit

$$
\sum_{i=0}^6 i y_i 
= 169 + 382 + 360 + 264 + 105 + 66
= 1346
\quad
\widehat{\lambda}
= \frac{1346}{650}
\approx 2,07.
$$

## 2. Erwartete Häufigkeiten unter \(H_0\)

Die Nullhypothese \(H_0\) lautet:
$$
\text{Die Anzahl solcher Zellen folgt einer Poisson-Verteilung mit Parameter } \lambda = \widehat{\lambda} \approx 2,07 
$$

Für die Poisson-Verteilung gilt:
$$
P(X = i) = \frac{e^{-\lambda}\,\lambda^i}{i!}.
$$
Die (engl. expected frequency) in Kategorie $i$ ist daher:
$$
E_i 
= 650 \times P(X = i)
= 650 \times \frac{e^{-\widehat{\lambda}}\,\widehat{\lambda}^i}{i!}.
$$

Zur Illustration einige näherungsweise Berechnungen (gerundete Werte):

$$E_0 \approx 650 \times e^{-2.07} \times 2.07^0 / 0! \approx 82$$

$$E_1 \approx 650 \times e^{-2.07} \times 2.07^1 / 1! \approx 170$$

$$E_2 \approx 650 \times e^{-2.07} \times 2.07^2 / 2! \approx 177$$

$$E_3 \approx 650 \times e^{-2.07} \times 2.07^3 / 3! \approx 121$$

$$E_4 \approx 650 \times e^{-2.07} \times 2.07^4 / 4! \approx 63$$

$$E_5 \approx 650 \times e^{-2.07} \times 2.07^5 / 5! \approx 26$$

$$E_6 \approx 650 \times e^{-2.07} \times 2.07^6 / 6! \approx 9$$

(Die Summe kann auf Grund von Rundungen etwa 648–650 ergeben.)

## 3. Teststatistik des $\chi^2$-Anpassungstests

Die $\chi^2$-Teststatistik lautet:

$$
\chi^2 
= \sum_{i=0}^{6} \frac{(y_i - E_i)^2}{E_i}.
$$

Beispielhafte (gerundete) Einzelbeiträge:
$$i=0: (72 - 82)^2 / 82 = 100/82 \approx 1,22$$

$$i=1: (169 - 170)^2 / 170 = 1/170 \approx 0,01$$

$$i=2: (191 - 177)^2 / 177 = 196/177 \approx 1,11$$

$$i=3: (120 - 121)^2 / 121 = 1/121 \approx 0,01$$

$$i=4: (66 - 63)^2 / 63 = 9/63 \approx 0,14$$

$$i=5: (21 - 26)^2 / 26 = 25/26 \approx 0,96$$

$$i=6: (11 - 9)^2 / 9 = 4/9 \approx 0,44$$

Summiert man diese Werte, erhält man ungefähr:

$$
\chi^2 
\approx 1,22 + 0,01 + 1,11 + 0,01 + 0,14 + 0,96 + 0,44 
\approx 3,88
$$

## 4. Freiheitsgrade und Entscheidung

Da wir $k=7$ Kategorien $i=0$ bis $i=6$ haben und $\lambda$ aus den Daten geschätzt wurde (1 zusätzlicher Parameter), gilt die Anzahl der Freiheitsgrade:

$$
\text{df} 
= k - 1 - \text{(Anzahl geschätzter Parameter)}
= 7 - 1 - 1 
= 5
$$

Auf dem 5%-Niveau ist der kritische $\chi^2$-Wert:

$$
\chi^2_{0.05,(5)} 
\approx 11,07.
$$

Da

$$
\chi^2_\text{beob} 
\approx 3,88 
< 11,07
$$
liegt der p-Wert über 5 %. Wir **verwerfen** die Nullhypothese $\bigl(H_0):$ Poisson($\lambda\approx2,07$ $\bigr)$ **nicht**.

## 5. Ergebnis

Die Daten sind **vereinbar** mit einer Poisson-Verteilung (Parameter $\widehat{\lambda} \approx 2,07$). Es gibt **keinen** statistisch signifikanten Hinweis auf eine Abweichung vom Poisson-Modell bei $\alpha = 5%$. 


### Aufgabe 4

Kann die Milchleistung \[in Hektolitern pro Jahr\] von Milchkühen einer bestimmten Züchtung durch eine normalverteilte Zufallsvariable mit $\mu=34$ und $\sigma=5$ angemessen beschrieben werden? Man untersuche die Frage mit dem $\chi^2$-Anpassungstest zum Niveau $\alpha=5\%$ aufgrund folgender Beobachtung:

$$
\begin{array}{|l|c|c|c|c|c|c|c|c|c|}
  \hline
  % after \\: \hline or \cline{col1-col2} \cline{col3-col4} ...
  \textrm{Milchleistung} & <28 & 28-30 & 30-32 & 32-34 & 34-36 & 36-38 & 38-40 & 40-42 & >42 \\ \hline
  \textrm{Anzahl der Kühe} & 11 & 13 & 16 & 19 & 22 & 18 & 15 & 6 & 5 \\
  \hline
\end{array}
$$

**Chi-Quadrat-Anpassungstest zum Niveau $\alpha = 5%$**  

Wir haben die Milchleistungen (in Hektolitern pro Jahr) einer Stichprobe von Kühen folgendermaßen klassiert:

$$
\begin{array}{c|ccccccccc}
\text{Klasse (Intervall)} 
 & <28 & 28-30 & 30-32 & 32-34 & 34-36 & 36-38 
 & 38-40 & 40-42 & >42 \\
\hline
\text{Beobachtet }(O_i) 
 & 11 & 13 & 16 & 19 & 22 & 18 & 15 & 6 & 5 
\end{array}
$$

Die **Summe** der beobachteten Häufigkeiten ist
$$
\sum O_i 
= 11 + 13 + 16 + 19 + 22 + 18 + 15 + 6 + 5 
= 125.
$$

Wir testen, ob diese Beobachtungen mit einer **Normalverteilung** $N(\mu=34,\sigma=5)$ verträglich sind. Unter der Nullhypothese $H_0$:  
Die Milchleistung folgt $N(34,5)$

## 1. Bestimmung der (theoretischen) Klassen-Wahrscheinlichkeiten

Wir berechnen für jede Klasse die **theoretische Wahrscheinlichkeit** unter der Normalverteilung $N(34,5)$. Zu den Klassen gehören Intervalle der Form $[a,b)$ (bis auf die untere Klasse $(-\infty,28)$ und die obere Klasse $(42,\infty)$.  

Bezeichne die Verteilungsfunktion der Standardnormalverteilung mit $\Phi(z)$. Dann gilt für ein Intervall $[a,b)$:

$$
P(a \le X < b)
= \Phi\!\Bigl(\frac{b - 34}{5}\Bigr)
 - \Phi\!\Bigl(\frac{a - 34}{5}\Bigr).
$$

Klasse $< 28$:  
  $\displaystyle P(X<28) 
  = \Phi\!\bigl(\tfrac{28 - 34}{5}\bigr)
  = \Phi(-1.2)$

Klasse $28-30$:  
  $\displaystyle P(28 \le X < 30) 
  = \Phi\!\bigl(\tfrac{30-34}{5}\bigr)
  - \Phi\!\bigl(\tfrac{28-34}{5}\bigr)
  = \Phi(-0.8) - \Phi(-1.2)$

Klasse $30{-}32$, usw.  
  Analog berechnet man für jedes Intervall die Differenz $\Phi(z_\text{oben}) - \Phi(z_\text{unten})$

Klasse $>42$:  
  $\displaystyle P(X>42) 
  = 1 - \Phi\!\bigl(\tfrac{42 - 34}{5}\bigr)
  = 1 - \Phi(1.6).$

Multipliziert man jede dieser Wahrscheinlichkeiten mit der Gesamtzahl $125$, erhält man die **erwarteten Häufigkeiten** $E_i$.  

## 2. Beispielhafte Berechnung der erwarteten Häufigkeiten

Mit ungefähren Standardnormalwerten (gerundete Werte):
$$\Phi(-1.2) \approx 0.1151$$

$$\Phi(-0.8) \approx 0.2119$$

$$\Phi(-0.4) \approx 0.3446$$

$$\Phi(0) = 0.5$$

$$\Phi(0.4) \approx 0.6554$$

$$\Phi(0.8) \approx 0.7881$$

$$\Phi(1.2) \approx 0.8849$$

$$\Phi(1.6) \approx 0.9452$$

lassen sich die Wahrscheinlichkeiten (und damit die erwarteten Häufigkeiten) näherungsweise bestimmen:

1. $P(X<28) = \Phi(-1.2) \approx 0.1151 
\Rightarrow E_1 \approx 125 \cdot 0.1151 \approx 14.39$

2. $P(28\le X<30) = \Phi(-0.8) - \Phi(-1.2) 
 \approx 0.2119 - 0.1151 = 0.0968 
 \Rightarrow E_2 \approx 125 \cdot 0.0968 \approx 12.1$

3. $P(30\le X<32) = \Phi(-0.4) - \Phi(-0.8) 
 \approx 0.3446 - 0.2119 = 0.1327 
 \Rightarrow E_3 \approx 125 \cdot 0.1327 \approx 16.6$

4. $P(32\le X<34) = \Phi(0) - \Phi(-0.4) 
 \approx 0.5 - 0.3446 = 0.1554 
 \Rightarrow E_4 \approx 125 \cdot 0.1554 \approx 19.4$

5. $P(34\le X<36) = \Phi(0.4) - \Phi(0) 
 \approx 0.6554 - 0.5 = 0.1554 
 \Rightarrow E_5 \approx 19.4$ (symmetrisch wie Interval $[32,34)$)

6. $P(36\le X<38) = \Phi(0.8) - \Phi(0.4) 
 \approx 0.7881 - 0.6554 = 0.1327 
 \Rightarrow E_6 \approx 16.6.$

7. $P(38\le X<40) = \Phi(1.2) - \Phi(0.8) 
 \approx 0.8849 - 0.7881 = 0.0968 
 \Rightarrow E_7 \approx 12.1$

8. $P(40\le X<42) = \Phi(1.6) - \Phi(1.2) 
 \approx 0.9452 - 0.8849 = 0.0603 
 \Rightarrow E_8 \approx 7.54$

9. $P(X>42) = 1 - \Phi(1.6) 
 \approx 0.0548 
 \Rightarrow E_9 \approx 6.85$

## 3. Berechnung der Teststatistik

Die **$\chi^2$-Teststatistik** lautet:

$$
\chi^2 
= \sum_{i=1}^{9} \frac{(O_i - E_i)^2}{E_i}.
$$

Beispielsweise für das erste Intervall $<28$:
$$
\frac{(11 - 14.39)^2}{14.39} \approx 0.80.
$$
Man verfährt so für alle 9 Klassen und summiert die Einzelbeiträge. Eine beispielhafte Aufsummierung ergibt in etwa:

$\chi^2 \approx 2.9$ bis $3.0$ (abhängig von Rundungen).

## 4. Freiheitsgrade und kritischer Wert

Da wir **9 Klassen** haben und **zwei Verteilungsparameter** $\mu$ und $\sigma$ vorgegeben sind (aber hier nicht aus den Daten geschätzt wurden), ist meist folgender Freiheitsgrad-Ansatz üblich:

$$
\text{df} 
= \bigl(\text{Anzahl Klassen}\bigr) - 1 
 - \bigl(\text{Anzahl geschätzter Parameter}\bigr).
$$

Bei Vorliegen **gegebener** (nicht aus den Daten geschätzter) Parameter kann man - je nach Lehrbuch - entweder $df = 9-1=8$ nehmen oder $df = 9-1-2=6$, sofern man streng nach „Parameter der Verteilung“ abzählt. Oft wird jedoch empfohlen, bei Anpassung an eine Normalverteilung **zwei** Freiheitsgrade für $\mu,\sigma$ abzuziehen.  

- Variante 1 (Parameter als *nicht* geschätzt):  
  $df = 9 - 1 = 8$  
- Variante 2 (Parameter formal abziehen):  
  $df = 9 - 1 - 2 = 6$

Auf dem 5 %-Niveau liegt der kritische Wert  
- für $df = 8$ etwa $\chi^2_{0.05;(8)} \approx 15.51$,  
- für $df = 6$ etwa $\chi^2_{0.05;(6)} \approx 12.59$.

In **beiden** Fällen gilt:

$$
\chi^2_\text{beob} \approx 3 < 12.59 \text{(bzw. 15.51)}
$$

Wir verwerfen $H_0$ folglich **nicht**.

## 5. Ergebnis

Die beobachtete $\chi^2$-Teststatistik liegt deutlich unter dem kritischen Wert, sodass **kein** statistisch signifikanter Widerspruch zur Normalverteilung $N(34,5)$ gefunden wird. Auf dem 5 %-Niveau sind die Daten **vereinbar** mit der Normalverteilungsannahme (Mittelwert 34, Standardabweichung 5).
