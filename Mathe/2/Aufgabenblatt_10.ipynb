### Aufgabe 6

Untersuchen Sie folgende Funktionen auf Extrema und Sattelpunkte:

a) $f(x,y)=x^2+y^2, D_f=\mathbb{R}^2$  
b) $f(x,y)=x^2-y^2, D_f=\mathbb{R}^2$

Plotten Sie die Funktionen.

a) $f(x,y)=x^2+y^2, D_f=\mathbb{R}^2$  

$$\nabla f = \left( \begin{array}{c}
                                      2x\\
                                      2y\\
                                    \end{array} \right) = 0$$
                                    $$x=0;y=0$$

$$H_f = \left( \begin{array}{ccc}
                                      2 & 0 \\
                                      0 & 2\\
                                        \end{array} \right)$$
                                        

$$det(H_f) = 4 > 0 ; \frac{\delta ^2 f}{\delta x^2} = 2> 0$$

(0;0) - Lokale Min Punkt (zumind.)

Die Funktion $f(x,y)=x^2+^2$ ist eine parabolische Oberfläche mit ihrem Scheitelpunkt im Punkt $(0,0)$. Somit wird das globale Minimum der Funktion $f(x,y)=x2+y2$ am Punkt $(0,0)$ erreicht und ist gleich $0$. Diese Funktion hat kein globales Maximum, ist also nicht von oben begrenzt und strebt gegen Unendlich, wenn es sich vom Punkt $(0,0)$ entfernt.

b) $f(x,y)=x^2-y^2, D_f=\mathbb{R}^2$  

$$\nabla f = \left( \begin{array}{c}
                                      2x\\
                                      -2y\\
                                    \end{array} \right) = 0$$
                                    $$x=0;y=0$$

$$H_f = \left( \begin{array}{ccc}
                                      2 & 0 \\
                                      0 & -2\\
                                        \end{array} \right)$$
                                        

$$det(H_f) = -4 < 0$$

(0;0) - Sattelpunkt

```python
from matplotlib import pyplot as plt
import numpy as np

def function(x, y):
    return x**2+y**2
 
 
x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
 
X, Y = np.meshgrid(x, y)
Z = function(X, Y)
 
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
 
ax.plot_surface(X, Y, Z, alpha=0.6)
 
 
plt.show()
```

$$3)(\pm\frac{3\sqrt{2}}{4};2) $$

$$H_f = \left( \begin{array}{ccc}
                                      0 & \pm 3\sqrt{2} \\
                                      \pm 3\sqrt{2} & -3\\
                                        \end{array} \right)$$

$$det(H)=0$$

Brauchen wir noch Analyse:

$$f(\pm \frac{3 \sqrt{2}}{4};\frac{1}{2}) = \frac{3}{8}$$

Bestimmen wir jetzt:

$$f(\pm \frac{3 \sqrt{2}}{4} + \epsilon;\frac{1}{2})$$- ist trotzdem $\frac{3}{8}$

$$f(\pm \frac{3 \sqrt{2}}{4};\frac{1}{2}+ \epsilon) = (\frac{1}{2}+\epsilon)^3+2(\pm\frac{3 \sqrt{2}}{4})^2(\frac{1}{2}+ \epsilon) - (\pm \frac{3 \sqrt{2}}{4})^2 = \epsilon^3-1.5\epsilon-\frac{9}{8}\epsilon+\frac{1}{2}$$

Hier sehen wir, dass der Wert der Funktion von $\epsilon$ abhängt und somit das Verhalten in der Umgebung variiert, was ein Eigenschaft von Sattelpunkt ist

$$3)(\pm\frac{3\sqrt{2}}{4};2) $$ - Sattelpunkte

```python
from matplotlib import pyplot as plt
import numpy as np

def function(x, y):
    return (y**3)+((2)*(x**2)*y)-(x**2)-(3)*(y**2)+1
 
 
x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
 
X, Y = np.meshgrid(x, y)
Z = function(X, Y)
 
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
 
ax.plot_surface(X, Y, Z, alpha=0.6)
 
 
plt.show()
```
