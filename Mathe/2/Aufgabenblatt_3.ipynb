### Aufgabe 1

Es sei $f(t)=\left( \begin{array}{c}
 \sin{t} \\
 \cos{t} \\
 t
  \end{array} \right)$. Plotten Sie die Funktion. Bestimmen Sie $\frac{df(t)}{dt},
  \frac{d^2f(t)}{dt^2}, \|\frac{df(t)}{dt}\|,\|
  \frac{d^2f(t)}{dt^2}\|$. 


```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.figure().add_subplot(projection = '3d')

t = np.linspace(0, 10*np.pi, 1000)
x = np.sin(t)
y = np.cos(t)
z = t

ax.plot(x, y, z)

ax.set_xlabel('X', fontweight = 'bold', fontsize = 14)
ax.set_ylabel('Y', fontweight = 'bold', fontsize = 14)
ax.set_zlabel('Z', fontweight = 'bold', fontsize = 14)

plt.title('Funktion f(t)', fontweight = 'bold', fontsize = 16)

plt.show()
```

$$\frac{df(t)}{dt}$$
$$f'(t)=\left( \begin{array}{c}\cos{t} \\ -\sin{t} \\ 1  \end{array} \right)$$


$$\frac{d^2f(t)}{dt^2}$$
$$f''(t)=\left( \begin{array}{c}-\sin{t} \\ -\cos{t} \\ 0  \end{array} \right)$$

$$\|\frac{df(t)}{dt}\| = \sqrt{\cos^2{t}+\sin^2{t}+1} = \sqrt{2}$$
$$\|\frac{d^2f(t)}{dt^2}\| = \sqrt{\sin^2{t}+\cos^2{t}} = 1$$
