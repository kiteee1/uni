### Aufgabe 1

Berechnen Sie die Fläche zwischen den Funktionen $f$ und $g$ im
Intervall zwischen den reellwertigen Schnittpunkten beider
Funktionen:

$$
f(x)=x^2,g(x)=\sqrt{x}
$$

Plotten Sie die beiden Funktionen.

Zuerst finden wir Schnittpunkten mittels:
$$x^2=\sqrt{x}$$

$$x=0$$
und
$$x=1$$

Flaeche:

$$\int_{0}^{1}\sqrt{x}-x^2dx=\frac{2x^{\frac{3}{2}}}{3}-\frac{x^3}{3}=\frac{2x\sqrt{x}-x^3}{3}\bigg|_{0}^{1}$$

$$\frac{2*1\sqrt{1}-1^3}{3}-\frac{2*0\sqrt{0}-0^3}{3}=\frac{1}{3}$$

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x):
   return x**2
def g(x):
    return np.sqrt(x)

x = np.linspace(-1, 2, 1000)
plt.ylim(-1, 3)
plt.plot(x, f(x), color='blue')
plt.plot(x, g(x), color='red')
plt.fill_between(x, f(x), g(x), where=g(x)>=f(x), color='grey')
plt.axhline()
plt.axvline()
plt.axhline(color='black', lw=0.5)
plt.axvline(color='black', lw=0.5)

plt.grid()
plt.show()
```

### Aufgabe 2

Berechnen Sie die Fläche zwischen den Funktionen $f$ und $g$ im
Intervall zwischen den reellwertigen Schnittpunkten beider
Funktionen:

$$
f(x)=4x,g(x)=x^2+2x-3
$$

Plotten Sie die beiden Funktionen.

Zuerst finden wir Schnittpunkten mittels:
$$x^2+2x-3=4x$$
$$x_{1} = -1, x_{2}=3$$

Flaeche

$$\int_{1}^{3}4x-(x^2-2x+3)dx=\int_{1}^{3}2x-x^2+3dx$$

$$\int 2xdx - \int x^2 dx+ \int 3dx = x^2 - \frac{x^3}{3}+3x$$

$$x^2 - \frac{x^3}{3}+3x\bigg|_{0}^{1}$$

$$3^2 - \frac{3^3}{3}+3*3-((-1)^2 - \frac{(-1)^3}{3}+3*(-1))=\frac{32}{3}$$

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x):
   return 4*x
def g(x):
    return x**2+2*x-3

x = np.linspace(-5, 5, 1000)
plt.ylim(-5, 15)
plt.plot(x, f(x), color='blue')
plt.plot(x, g(x), color='red')
plt.fill_between(x, f(x), g(x), where=f(x)>=g(x), color='grey')
plt.axhline()
plt.axvline()
plt.axhline(color='black', lw=0.5)
plt.axvline(color='black', lw=0.5)

plt.grid()
plt.show()
```

### Aufgabe 3

Berechnen Sie die Fläche zwischen den Funktionen $f$ und $g$ im
Intervall zwischen den reellwertigen Schnittpunkten beider
Funktionen:

$$
f(x)=\log{x},g(x)=(\log{x})^2
$$

Plotten Sie die beiden Funktionen.

Zuerst finden wir Schnittpunkten mittels:
$$\log{x}=\log^2{x}$$
$$x_{1} = -1, x_{2}=3$$
Lass $\log{x}=t$, dann
$$t=t^2$$
$$t=0, t=1$$
$$x=1, x=e$$

Flaeche

$$\int_{1}^{3}\log{x}-\log^2{x}dx=\int_{1}^{3}\log{x}dx - \int_{1}^{3}\log^2{x}dx$$

$$\int\log{x}dx = x\log{x}-x$$

$$\int\log^2{x}dx$$\
$$u = \log^2{x}, du = 2\log{x}*\frac{1}{x}dx$$
$$dv = dx, v=x$$
$$\int udv = uv-\int vdu$$
$$x\log^2{x} - \int x*2\log{x}*\frac{1}{x}dx = x\log^2{x} - 2(x\log{x}-x)=x\log^2{x}-2x\log{x}+2x$$

$$x\log{x}-x-(x\log^2{x}-2x\log{x}+2) = 3x\log{x}-3x-x\log^2{x}$$

$$3x\log{x}-3x-x\log^2{x}\bigg|_{1}^{e}$$

$$3*e\log{e}-3*e-e*\log^2{e}-(3*1\log{1}-3*1-1*\log^2{1}) = 3 - e$$

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x):
   return np.log(x)
def g(x):
    return np.log(x)**2

x = np.linspace(0, 5, 1000)
plt.ylim(-1, 5)
plt.plot(x, f(x), color='blue')
plt.plot(x, g(x), color='red')
plt.fill_between(x, f(x), g(x), where=f(x)>=g(x), color='grey')
plt.axhline()
plt.axvline()
plt.axhline(color='black', lw=0.5)
plt.axvline(color='black', lw=0.5)

plt.grid()
plt.show()
```

### Aufgabe 4

Dreht man eine Kurve mit $y=f(x)$ um die $x$-Achse, so beträgt das
Volumen des entstehenden Drehkörpers:

$$V=\pi\int_a^b{y^2dx}$$

Die Kurve mit $y=\cosh{x}$ rotiert um die $x$-Achse. Berechnen Sie
das Volumen des Drehkörpers zwischen $a=0$ und $b=2$.

Plotten Sie die Kurve.

$$2\cosh^2{x}-1=2(\frac{e^x-e^{-x}}{2})^2-1=\frac{(e^{x})^2+2e^xe^{-x}+(e^{-x})^2}{2}-1=\frac{e^{2x}+2+e^{-2x}-2}{2}=\frac{e^{2x}+e^{-2x}}{2}=\cosh{2x}$$
also $\cosh^2{x}=\frac{\cosh{2x}+1}{2}$

$$\pi\int_{0}^{2}\cosh^2{x}dx$$
$$\int\frac{\cosh{2x}+1}{2}dx = \frac{1}{2}\int\cosh{2x}+1 = \frac{1}{2}(\int \cosh{2x}dx +\int1dx)$$

$$\int \cosh{2x}dx$$
lass $t=2x$ dann
$$\int \frac{\cosh{t}}{2}dt = \frac{1}{2}\int\cosh{t}dt = \frac{1}{2}*\sinh{t}=\frac{\sinh{2x}}{2}$$

$$\frac{1}{2}(\int \cosh{2x}dx +\int1dx) = \frac{\sinh{2x}}{4}+\frac{1}{2}x$$

$$\frac{\sinh{2x}}{4}+\frac{1}{2}x\bigg|_{0}^{2}$$

$$\frac{\sinh{2*2}}{4}+\frac{1}{2}*2-(\frac{\sinh{2*0}}{4}+\frac{1}{2}*0) = \frac{e^4-e^{-4}}{8}+1$$

$$(\frac{e^4-e^{-4}}{8}+1)\pi\approx 24.575$$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

n = 1000

fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

z = np.linspace(-3, 3, n)
y = np.cosh(z)

theta = np.linspace(0, 2 * np.pi, n)

yn = np.outer(y, np.cos(theta))
zn = np.outer(y, np.sin(theta))
xn = np.zeros_like(yn)

for i in range(len(z)):
    xn[i:i + 1, :] = np.full_like(xn[0, :], z[i])

ax1.set_ylim(-10, 10)

ax1.axhline()
ax1.axvline()
ax1.axhline(color='black', lw=0.5)
ax1.axvline(color='black', lw=0.5)
ax1.grid()

ax1.plot(z, y, color='blue')

ax2.plot_surface(xn, yn, zn, alpha=0.7)

plt.show()
```

### Aufgabe 5

Dreht man eine Kurve mit $y=f(x)$ um die $y$-Achse, so beträgt das
Volumen des entstehenden Drehkörpers:

$$V=\pi\int_{y_1}^{y_2}{x^2dy}$$

Die Hyperbel $\frac{x^2}{a^2}-\frac{y^2}{b^2}=1$ rotiert um die
$y$-Achse. Berechnen Sie das Volumen des Drehkörpers zwischen
$y_1=-c$ und $y_2=c$.

Plotten Sie die Hyperbel.

$$x^2=a^2+\frac{y^2a^2}{b^2}$$

$$\int a^2+\frac{a^2y^2}{b^2}dy = \int a^2dy + \int\frac{a^2y^2}{b^2}dy = a^2y+\frac{a^2}{b^2}*\frac{y^3}{3}$$

$$a^2y+\frac{a^2}{b^2}*\frac{y^3}{3}\bigg|_{-c}^{c}$$

$$a^2c+\frac{a^2}{b^2}*\frac{c^3}{3}-(a^2(-c)+\frac{a^2}{b^2}*\frac{(-c)^3}{3})=\frac{6a^2b^2c+2a^2c^3}{3b^2}$$

$$V=\pi\frac{6a^2b^2c+2a^2c^3}{3b^2}$$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

n = 1000

fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122,projection='3d')
y = np.linspace(-10, 10, n)
x = 3*np.sqrt(1+((y**2)/(9)))
t = np.linspace(0, np.pi*2, n)
x1 = -3*np.sqrt(1+((y**2)/(9)))

xn = np.outer(x, np.cos(t))
yn = np.outer(x, np.sin(t))
zn = np.zeros_like(xn)

for i in range(len(x)):
    zn[i:i+1,:] = np.full_like(zn[0,:], y[i])

ax1.axhline()
ax1.axvline()
ax1.axhline(color='black', lw=0.5)
ax1.axvline(color='black', lw=0.5)
ax1.grid()

ax1.plot(x, y, color='blue')
ax1.plot(x1, y, color='blue')
ax2.plot_surface(xn, yn, zn)
plt.show()
```

### Aufgabe 6

Die Bogenlänge $s$ eines Stücks der Kurve mit $y=f(x)$ zwischen
$x=a$ und $x=b$ ergibt sich aus:

$$s=\int_{a}^{b}{\sqrt{1+y'^2}dx}$$

Berechnen Sie den Umfang der gleichseitigen Astroide, deren
Gleichung $x^{2/3}+y^{2/3}=r^{2/3}$ lautet.

Plotten Sie die Astroide.

$$y=\pm(r^{\frac{2}{3}}-x^{\frac{2}{3}})^{\frac{3}{2}}$$

Nehmen wir nur positive Teil

$$y'=\frac{3}{2}(r^{\frac{2}{3}}-x^{\frac{2}{3}})^{\frac{1}{2}}*(-\frac{2}{3}x^{-\frac{2}{3}}) = -\frac{(r^{\frac{2}{3}}-x^{\frac{2}{3}})^{\frac{1}{2}}}{x^{\frac{1}{3}}}$$

$$y'^2=\frac{r^{\frac{2}{3}}-x^{\frac{2}{3}}}{x^{\frac{2}{3}}}$$

Denn wir haben nur positive teil genommen, mussen wir Integral verdoppeln

$$s=2\int \sqrt{1+\frac{r^{\frac{2}{3}}-x^{\frac{2}{3}}}{x^{\frac{2}{3}}}} = 2\int \sqrt{\frac{x^{\frac{2}{3}}+r^{\frac{2}{3}}-x^{\frac{2}{3}}}{x^{\frac{2}{3}}}}dx = 2\int \sqrt{\frac{r^{\frac{2}{3}}}{x^{\frac{2}{3}}}}dx = 2\int \frac{r^{\frac{1}{3}}}{x^{\frac{1}{3}}}dx=2\sqrt[3]{r}\int \frac{1}{x^{\frac{1}{3}}}dx = 2\sqrt[3]{r}*\frac{3\sqrt[3]{x^2}}{2}=3\sqrt[3]{rx^2}$$

$$3\sqrt[3]{rx^2}\bigg|_{a}^{b}$$

$$3\sqrt[3]{ra^2}-3\sqrt[3]{rb^2}$$

### Aufgabe 7

Dreht man eine Kurve mit $y=f(x)$ um die $x$-Achse, so ist die
Mantelfläche des entstehenden Drehkörpers:

$$M=2\pi\int_a^b{y\sqrt{1+y'^2}dx}$$

Eine kubische Parabel mit $y=x^3$ rotiert um die $x$-Achse. Wie groß
ist die Mantelfläche für $a=0$ und $b=1$?

Plotten Sie die Parabel.

$$y'=3x^2$$

$$M=2\pi\int_0^1{x^3\sqrt{1+(3x^2)^2}dx}$$

$$\int{x^3\sqrt{1+9x^4}dx}$$
Lass $t=1+9x^4$, dann
$$dx = \frac{1}{t'}dt$$
$$t' = 36x^3$$
$$\int{x^3\sqrt{1+9x^4}\frac{1}{36x^3}dt} = \frac{1}{36}\int\sqrt{t}dt = \frac{1}{36}*\frac{2t\sqrt{t}}{3} = \frac{(1+9x^4)\sqrt{1+9x^4}}{54}$$

$$\frac{(1+9x^4)\sqrt{1+9x^4}}{54}\bigg|_0^1$$

$$\frac{(1+9(1)^4)\sqrt{1+9(1)^4}}{54}-(\frac{(1+9(0)^4)\sqrt{1+9(0)^4}}{54}) = \frac{5\sqrt{10}}{27}-\frac{1}{54}$$

$$M=2\pi(\frac{5\sqrt{10}}{27}-\frac{1}{54})\approx 3.563$$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

n = 1000

fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

z = np.linspace(-2, 2, n)
y = z**3

theta = np.linspace(0, 2 * np.pi, n)

yn = np.outer(y, np.cos(theta))
zn = np.outer(y, np.sin(theta))
xn = np.zeros_like(yn)

for i in range(len(z)):
    xn[i:i + 1, :] = np.full_like(xn[0, :], z[i])

ax1.set_ylim(-5, 5)

ax1.axhline()
ax1.axvline()
ax1.axhline(color='black', lw=0.5)
ax1.axvline(color='black', lw=0.5)
ax1.grid()

ax1.plot(z, y, color='blue')

ax2.plot_surface(xn, yn, zn, alpha=0.7)

plt.show()
```

### Aufgabe 8


Die Flächenschwerpunktskoordinaten $x_0,y_0$ einer Fläche $A$
ergeben sich aus den statischen Momenten $M_y,M_x$:

$$A=\int_a^bydx, M_y=\int_a^bxydx, M_x=\frac{1}{2}\int_a^by^2dx, x_0=\frac{M_y}{A}, y_0=\frac{M_x}{A}$$

Berechnen Sie die Flächenschwerpunktskoordinaten für einen Halbkreis
mit $M(0|0)$ und $y\geq0$.

Plotten Sie den Halbkreis.

Gleichung eines Halbkreises mit Radius r in der oberen Halbebene: $y=\sqrt{r^2-x^2}$ 

Unter der Annahme, dass$ M(0|0)$ der Mittelpunkt eines Kreises ist, können wir sagen, dass $a = -r$ und$ b = r$

$$A=\int_{-r}^{r} \sqrt{r^2-x^2}dx$$
$$\int\sqrt{r^2-x^2}dx$$
Sei $$x =r\sin{t}$$, dann
$$dx = x'dt$$
$$x' = r\cos{t}$$
$$\int\sqrt{r^2-(r\sin{t})^2}r\cos{t}dt$$
$$r\int r\sqrt{1-\sin{t}^2}\cos{t}dt$$
$$r\int r\sqrt{\cos^2{t}}\cos{t}dt$$
$$r\int r\cos^2{t}dt$$
$$r^2\int\frac{1+cos{2t}}{2}dt$$
$$\frac{1}{2}r^2\int 1+cos{2t}dt = \frac{1}{2}r^2(\int 1 dt +\int cos{2t} dt)$$
$$\int cos{2t}dt$$
Sei $u = 2t$, dann
$$\int\frac{\cos{u}}{2}du = \frac{1}{2}\sin{u} = \frac{\sin{2t}}{2}$$
$$\frac{1}{2}r^2(\int 1 dt +\int cos{2t} dt) = \frac{1}{2}r^2(t +\frac{\sin{2t}}{2})$$
$$t = \arcsin{\frac{x}{r}}$$
$$\frac{1}{2}r^2(\arcsin{\frac{x}{r}} +\frac{\sin{(2*\arcsin{\frac{x}{r}})}}{2}) = \frac{r^2\arcsin{(\frac{x}{r})}}{2} + \frac{r^2*2\sin{(\arcsin{(\frac{x}{r})})}\cos{(\arcsin{(\frac{x}{r})})}}{4}=\frac{r^2\arcsin{(\frac{x}{r})}}{2} + \frac{r^2*2\frac{x}{r}\sqrt{1-(\frac{x}{r})^2}}{4} = \frac{r^2\arcsin{(\frac{x}{r})}}{2} + \frac{rx\sqrt{\frac{r^2-x^2}{r^2}}}{2}=\frac{r^2\arcsin{(\frac{x}{r})}}{2} + \frac{x\sqrt{r^2-x^2}}{2}$$
$$\frac{r^2\arcsin{(\frac{x}{r})}+x\sqrt{r^2-x^2}}{2}\bigg|_{-r}^r$$
$$\frac{r^2\arcsin{(\frac{r}{r})}+r\sqrt{r^2-r^2}}{2}-(\frac{r^2\arcsin{(\frac{-r}{r})}+(-r)\sqrt{r^2-(-r)^2}}{2})=\frac{\pi r^2}{2}=A$$

$$M_y=\int_{-r}^r x\sqrt{r^2-x^2}dx$$

$$\int x\sqrt{r^2-x^2}dx$$

Sei $t = r^2-x^2$, dann
$$dx = \frac{1}{t'}dt$$
$$t' = -2x$$
$$\int x\sqrt{r^2-x^2}*\frac{1}{-2x}dt = -\frac{1}{2}\int\sqrt{t}dt = \frac{1}{2}*\frac{2t\sqrt{t}}{3}=-\frac{(r^2-x^2)\sqrt{r^2-x^2}}{3}$$

$$-\frac{(r^2-x^2)\sqrt{r^2-x^2}}{3}\bigg|_{-r}^r$$
$$M_y=-\frac{(r^2-r^2)\sqrt{r^2-r^2}}{3}+\frac{(r^2-(-r)^2)\sqrt{r^2-(-r)^2}}{3} =0$$

$$M_x=\frac{1}{2}\int_{-r}^r r^2-x^2dx$$

$$\int r^2-x^2dx = r^2x-\frac{x^3}{3}$$

$$r^2x-\frac{x^3}{3}\bigg|_{-r}^r$$

$$M_x = r^2*r-\frac{r^3}{3}-(r^2*(-r)-\frac{(-r)^3}{3}) = \frac{4}{3}r^3$$

$$x_0=\frac{M_y}{A}, y_0=\frac{M_x}{A}$$
$$x_0=\frac{0}{\frac{\pi r^2}{2}}, y_0=\frac{\frac{4}{3}r^3}{\frac{\pi r^2}{2}}$$
$$x_0=0, y_0=\frac{8r}{3\pi}$$
