2. Erzeugen Sie einen Vektor namens Groesse mit den Werten (S, M, L) und einen
weiteren Vektor namens Anzahl mit den Werten (130,210,170). <br>
a) Zeichnen Sie zwei zugehörige Histogramme, einmal mit den aboluten und mit den relativen Häufigkeiten.<br>
b) Ersetzen Sie die Labels (S, M, L) durch (34, 38, 42) und machen Sie das gleiche wie in a). <br>
c) Berechnen Sie drei statistische Kennwerte mithilfe Python Funktionen. <br>
d) Fügen Sie die beiden Variablen Groesse und Anzahl zu einem Data Frame namens Daten
zusammen.

a)

```python
groesse = np.array(['S', 'M', 'L'])
anzahl = np.array([130, 210, 170])

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(groesse, anzahl)
plt.title('Absolute Häufigkeiten')
plt.ylabel('Anzahl')

relative_frequencies = anzahl / np.sum(anzahl)
plt.subplot(1, 2, 2)
plt.bar(groesse, relative_frequencies)
plt.title('Relative Häufigkeiten')
plt.ylabel('Relative Anzahl')

plt.tight_layout()
plt.show()
```

b)

```python
groesse_replaced = np.array([34, 38, 42])

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(groesse_replaced, anzahl)
plt.title('Absolute Häufigkeiten (Ersetzte Labels)')
plt.ylabel('Anzahl')

plt.subplot(1, 2, 2)
plt.bar(groesse_replaced, relative_frequencies)
plt.title('Relative Häufigkeiten (Ersetzte Labels)')
plt.ylabel('Relative Anzahl')

plt.tight_layout()
plt.show()
```

c)

```python
mean_value = np.mean(anzahl)
median_value = np.median(anzahl)
std_dev_value = np.std(anzahl)

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Standard Deviation: {std_dev_value}")
```

d)

```python
daten = pd.DataFrame({
    'Groesse': groesse,
    'Anzahl': anzahl
})

print(daten)
```
